#**Importing the libraries**#

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#**Loading the Dataset**#

```python
#reading the csv file
cc_data=pd.read_csv("creditcard.csv",error_bad_lines=False)

```

```python
# first 5 rows of the data
cc_data.head()
```

```python
# Last 5 rows of the data
cc_data.tail()
```

```python
# dataset information
cc_data.info()
```

```python
# checking missing values in each column
cc_data.isnull().sum()
```

```python
# no. of legit and fraudlent transactions
cc_data['Class'].value_counts()
```

#**Seperating the data for analysis**#

```python
legit=cc_data[cc_data.Class==0]
fraud=cc_data[cc_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare values for both transactions
cc_data.groupby('Class').mean()
```

#**Creating a sample dataset with a similar proportion of normal and fraudlent transactions by under-sampling**#

```python
#Number of fraudlent transactions is 195

legit_sample=legit.sample(n=195)
```

#**Concatenating two DataFrames**#

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

#**Splitting the data into features and targets.**#

```python
X=new_dataset.drop(columns="Class",axis=1)
Y=new_dataset['Class']
print(X)
print(Y)
```

##**Splitting the data into training and testing Data**##

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

#**Model Training**#

```python
model=LogisticRegression()

# training the logistic regression model with training data
model.fit(X_train,Y_train)
```

#**Model Evaluation : Accuracy Score**#

```python
# accuracy on training data

X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data :",training_data_accuracy)
```

```python
# accuracy on test data

X_test_prediction = model.predict(X_test)
testing_data_accuracy= accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on testing data :",testing_data_accuracy)
```
