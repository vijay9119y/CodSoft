# **Importing the Libraries and Dataset.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as plb
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

titanic_data = pd.read_csv("train.csv")

#set the global size of matplotlib figure
plt.rc('figure', figsize=(10,5))

#set the size of my subplots
size_subplots=(10,10)

#size of histogram bins
bin_size = 10
```

#**Data Cleaning & Preprocessing**

```python

#printing first 5 rows of the dataset
titanic_data.head()
```

```python
#getting information about the data
titanic_data.info()
```

```python
#no. of rows and columns
titanic_data.shape
```

```python
# check no. of misssing values on each column
titanic_data.isnull().sum()
```

#Handling Missing Values

```python
# replacing the missing values in "Age" column with all their mean.
titanic_data['Age'].fillna(titanic_data['Age'].mean(),inplace=True)

# finding the mode value of the "Embarked" column and replacing the missing values
print(titanic_data['Embarked'].mode()[0])
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0],inplace=True)
```

```python
# dropping the "Cabin" column from the dataframe
titanic_data = titanic_data.drop(columns='Cabin',axis=1)
```

```python
# Again checking the no. of missing values
titanic_data.isnull().sum()
```

#**Data Analysis**

```python
# getting some statistical measures about the data
titanic_data.describe()
```

```python
# finding the number of people survived and not survived
titanic_data['Survived'].value_counts()
```

#**Data Visualization**

```python
fig = plb.figure(figsize=size_subplots)
fig_dims = (3,2)

plb.subplot2grid(fig_dims, (0,0))
titanic_data['Survived'].value_counts().plot(kind = 'bar', title = 'Death vs survival count')

plb.subplot2grid(fig_dims, (0,1))
titanic_data['Pclass'].value_counts().plot(kind = 'bar', title = 'Passenger class count')

plb.subplot2grid(fig_dims, (1,0))
titanic_data['Sex'].value_counts().plot(kind = 'bar', title = 'Gender Counts')

plb.xticks(rotation=0)

plb.subplot2grid(fig_dims, (1,1))
titanic_data['Embarked'].value_counts().plot(kind = 'bar', title = 'Embarked counts')

plb.subplot2grid(fig_dims, (2,0))
titanic_data['Age'].hist()
plb.title('AGE histogram')
```

#**Encoding the Categorical Column**

```python
titanic_data['Sex'].value_counts()
```

```python
titanic_data['Embarked'].value_counts()
```

```python
# converting categorical columns

titanic_data.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)
titanic_data.head()
```

#**Seperating features and Target column("Survived")**

```python
X=titanic_data.drop(columns=['PassengerId','Name','Ticket','Survived'],axis=1)
Y=titanic_data['Survived']
print(X)
```

```python
print(Y)
```

#**Spliting the data into training data and test data**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

#**Model Training(Logistic Regression Model)**

```python
model=LogisticRegression()
```

```python
# training the Logistic Regression Model with training data

model.fit(X_train,Y_train)
```

```python
LogisticRegression()
```

#**Model Evaluation : Accuracy Score**

```python
# accuracy on training data

X_train_prediction=model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy=accuracy_score(Y_train,X_train_prediction)
print("Accuracy Score of Training Data",training_data_accuracy)
```

```python
# accuracy on test data

X_test_prediction=model.predict(X_test)
print(X_test_prediction)
```

```python
test_data_accuracy=accuracy_score(Y_test,X_test_prediction)
print("Accuracy Score of Test Data",test_data_accuracy)
```
